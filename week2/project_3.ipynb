```python
# 1. Data Initialization
girls_names = ["Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa"]
girls_age = [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
girls_height = [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
girls_scores = [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]

boys_names = ["Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"]
boys_age = [19, 16, 18, 17, 20, 19, 16, 18, 17, 19]
boys_height = [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7]
boys_scores = [74, 87, 75, 68, 66, 78, 87, 98, 54, 60]

# 2. Combining Data
names = girls_names + boys_names
ages = girls_age + boys_age
heights = girls_height + boys_height
scores = girls_scores + boys_scores

# 3. Generating the Table
print(f"{'Name':<12} | {'Age':<5} | {'Height':<8} | {'Score':<5}")
print("-" * 40)

for i in range(len(names)):
    print(f"{names[i]:<12} | {ages[i]:<5} | {heights[i]:<8} | {scores[i]:<5}")
```

Initialize Data Structures: Create separate lists for the attributes (Names, Age, Height, Scores) for both girls and boys.

Merge Groups: Combine the girls' data and boys' data into a single collection to represent the entire class.

Define Table Headers: Set the column titles (Name, Age, Height, Score).

Iterate and Align: Loop through the lists simultaneously. Since the data is "respective," the element at index $i$ in the Name list corresponds to the element at index $i$ in the Age, Height, and Score lists.

Format Output: Print each row with consistent spacing or padding to ensure the columns align vertically.
