START
    IF cart IS EMPTY THEN
        DISPLAY "Your cart is empty. Please add items before checking out."
        RETURN
    END IF
    
    item_final_price = 0
    FOR EACH item IN cart:
        item_final_price = item_final_price + (item.price * item.quantity)
    ENDFOR

    tax = subtotal * 0.05  // Assuming 5% tax
    total_amount = subtotal + tax

    DISPLAY "Total to pay: " + total_amount

    payment_successful = FALSE
    WHILE payment_successful IS FALSE:
        payment_method = GET user_input("Select Payment Method (Card/Transfer/Digital Wallet)")
        
        payment_status = PROCESS_PAYMENT(user_id, total_amount, payment_method)

        IF payment_status == "SUCCESS" THEN
            payment_successful = TRUE
        ELSE
            DISPLAY "Payment failed. Please try again or change payment method."
            IF user_wants_to_cancel() THEN RETURN
        END IF
    END WHILE

    transaction_id = GENERATE_UNIQUE_ID()
    UPDATE item_stock(cart)
    SAVE_TRANSACTION_TO_DB(transaction_id, user_id, cart, total_amount)

    receipt = CREATE_RECEIPT(transaction_id, cart, total_amount)
    DISPLAY "Thank you for your purchase!"
    PRINT_OR_EMAIL(receipt)

   END
