```python
def calculate_atr():
    print("--- Izifin Technology ATR Calculator ---")
    
    try:
        # 1. Input gathering
        experience = int(input("Enter years of work experience: "))
        age = int(input("Enter age of the staff: "))

        # 2. Logic processing using nested conditions
        if experience > 25 and age >= 55:
            atr = 5_600_000
        elif experience > 20 and age >= 45:
            atr = 4_480_000
        elif experience > 10 and age >= 35:
            atr = 1_500_000
        else:
            # Default for staff with less than 10 years exp and below 35 age
            atr = 550_000

        # 3. Output display
        print(f"\nThe Annual Tax Revenue (ATR) for this staff is: N{atr:,}")

    except ValueError:
        print("Invalid input. Please enter numerical values for age and experience.")

# Run the program
calculate_atr()
```
