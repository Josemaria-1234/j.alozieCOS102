```python
import numpy as np

print("=== Simple Equation Solver ===")
print("1. Quadratic ($Ax^2 + Bx + C = 0$)")
print("2. Cubic ($Ax^3 + Bx^2 + Cx + D = 0$)")
print("3. Quartic ($Ax^4 + Bx^3 + Cx^2 + Dx + E = 0$)")

# Get the user's choice
choice = input("\nWhat do you want to solve? (Type 1, 2, or 3): ")

if choice == '1':
    print("\n-- Solving Quadratic --")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
   
    # Calculate and print roots
    roots = np.roots([a, b, c])
    print("\nThe roots are:")
    print(roots)

elif choice == '2':
    print("\n-- Solving Cubic --")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
   
    roots = np.roots([a, b, c, d])
    print("\nThe roots are:")
    print(roots)

elif choice == '3':
    print("\n-- Solving Quartic --")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))
   
    roots = np.roots([a, b, c, d, e])
    print("\nThe roots are:")
    print(roots)

else:
    print("\nInvalid choice. Please restart and pick 1, 2, or 3.")
```
